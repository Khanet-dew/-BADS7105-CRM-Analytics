```python
import pandas as pd

df = pd.read_csv('Customer.csv')
```

```python
df.head(5)
```

```python
df = df.fillna("ไม่เคยซื้อ")
missing = df.isnull().sum()
print(missing)
```

```python
df = df.replace(to_replace = ["ไม่เคยซื้อ","ไม่เคยซือ","ไม่เคย","ไม่","เคยซื้อ","เคย"], value = [0,0,0,0,1,1])
df = df.drop(columns=["Timestamp"])
df.head(5)
```

```python
df.dtypes
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
df = df.applymap(encode_units).astype(int)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
```

```python
rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.9) ]
rules.sort_values(['support'], ascending = False).head(20)
```

**สรุป**

การตั้งค่านั้นให้หา Support ขั้นต่ำคือ 0.3 และสร้างกฏเพื่อกำหนดค่า lift อยู่ที่มากกว่าหรือเท่ากับ 1 ส่วน confidence ต้องมีค่ามากกว่าหรือเท่ากับ 0.9

จากนั้นนำมาเรียงลำดับตามค่า Support เพื่อหา Top 20 ของการซื้อของนักเรียนในห้องจะเห็นได้ว่าแซลมอนกับยาดมนั้นเป็นดังสิ่งที่เยียวยาจิตใจนักศึกษาวิชานี้ให้รอดพ้นจากการเรียนอันนักหน่วง ถ้ามหาลัยจะแจกรางวัลคนเรียนดีควรให้แซลมอลและยาดมเพื่อเป็นแรงกระตุ้นให้นักศึกษามีกำลังใจในเรียนวิชาต่างๆ
